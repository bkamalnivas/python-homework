```python
import sys
import csv
from pathlib import Path

# print o/p to file
def print_to_file(output_str):
    #open txt file & write the string
    file_path = Path('output.txt');
    with open(file_path, 'w') as file_pointer:
        file_pointer.write(output_str);

    
# get the o/p string in desired format
def get_output_str (month_count, total, avg_change,
                    profit_max, profit_min):
    
    output_str = "Total Months: " + str(month_count) + "\n";
    output_str += "Total: " + str(total) + "\n";
    output_str += "Average Change: $" + str(round(avg_change, 2)) + "\n";
    output_str += "Greatest Increase in Profits: " + str(profit_max[0]) + "(" + str(profit_max[1]) + ")" + "\n";
    output_str += "Greatest Decrease in Profits: " + str(profit_min[0]) + "(" + str(profit_min[1]) + ")";
    
    return output_str;


# main function
def main():

    #Declare all vars
    budget_data = [];
    max_profit = -sys.maxsize - 1;
    min_profit = sys.maxsize;
    total, total_change, avg_change = 0, 0, 0;
    profit_max_index, profit_min_index = 0, 0;
    
    # read the CSV & process it as needed
    file_path = Path('budget_data.csv');
    with open(file_path) as file_pointer:
        csv_reader = csv.reader(file_pointer, delimiter=',');
        next(csv_reader, None);
        budget_data = list(csv_reader);
    
    # initialize the total with first entry
    total += int(budget_data[0][1]);
    
    # iterate from index = 1 to end of list
    for index in range(1, len(budget_data)):       
        # calculate totals
        total += int(budget_data[index][1]);
        profit = int(budget_data[index][1]) - int(budget_data[index - 1][1]);
        total_change += profit;
        
        #update max & min as needed
        if (profit > max_profit):
            max_profit = profit;
            profit_max_index = index;
            
        if (profit < min_profit):
            min_profit = profit;
            profit_min_index = index;
    
    #calulate avg chage & month count
    avg_change = total_change / (len(budget_data) - 1);
    month_count = len(budget_data);
    
    # get the o/p string in desired format
    output_str = get_output_str(len(budget_data), total, avg_change,
                                budget_data[profit_max_index], 
                                budget_data[profit_min_index]);
    
    #print(output_str);
    print_to_file(output_str);
    

main();
```
